```python
import matplotlib.pyplot as çizim
import math 

# Kullanıcıdan (x1, y1) ve (x2, y2) nokta girdilerini alın
x1 = int(float(input("X Başlangıç değerini girin: ")))
y1 = int(float(input("Y Başlangıç değerini girin: ")))
x2 = int(float(input("X Bitiş değerini girin: ")))
y2 = int(float(input("Y Bitiş değerini girin: ")))

#koordinatlar = input("Koordinatları girin (x1, y1, x2, y2): ").split(',')
#x1, y1, x2, y2 = map(float, koordinatlar)

# Formüller
Oklidyenuzaklik  = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
ManhattanUzaklik = abs(x1 - x2) + abs(y1 - y2)

a=abs(x1 - x2)
b=abs(y1 - y2)
MaksimumUzaklık = max(a,b)



# Noktaları çizmek için grafik oluşturun
çizim.figure()
çizim.plot([x1, x2], [y1, y2], marker='^', linestyle='-', color='b')
çizim.xlabel('x Ekseni')
çizim.ylabel('y Ekseni')

# A ve B noktalarını çizgi üzerine ekleyin
çizim.text(x1, y1, 'A', fontsize=12, ha='right', va='bottom')
çizim.text(x2, y2, 'B', fontsize=12, ha='right', va='bottom')


çizim.title(f"Uzaklık Fonksiyonlarına göre iki nokta arasındaki uzaklık") #başlık(title)
çizim.grid(False)  #kareler--False siler True ekler

# Grafik penceresini gösterin
çizim.show()
print(f" A1=({x1},{y1}) A2=({x2},{y2}) noktaları arasındaki Öklidyen uzaklık {Oklidyenuzaklik}")
print(f" A1=({x1},{y1}) A2=({x2},{y2}) noktaları arasındaki Manhattan uzaklık {ManhattanUzaklik}")
print(f" A1=({x1},{y1}) A2=({x2},{y2}) noktaları arasındaki Maksimum uzaklık : {MaksimumUzaklık}")
```
