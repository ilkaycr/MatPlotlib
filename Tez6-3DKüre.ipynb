```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Topun parametreleri
radius = 1

# Theta ve phi açıları
theta = np.linspace(0, 2*np.pi, 100)
phi = np.linspace(0, np.pi, 100)

# Theta ve phi'yi bir ızgara haline getir
theta, phi = np.meshgrid(theta, phi)

# Topun x, y, z koordinatları
x = radius * np.sin(phi) * np.cos(theta)
y = radius * np.sin(phi) * np.sin(theta)
z = radius * np.cos(phi)

# 3D figürü oluştur
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Topu çiz
ax.plot_surface(x, y, z, color='blue', alpha=0.7, edgecolor='k')

# Plot'u göster
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def draw_sphere(ax, radius=1, center=(0, 0, 0), color='b', alpha=0.6):
    # Küre çizimi için theta ve phi değerleri
    theta = np.linspace(0, 2 * np.pi, 100)
    phi = np.linspace(0, np.pi, 100)

    # Küre yüzey noktalarının hesaplanması
    x = center[0] + radius * np.outer(np.cos(theta), np.sin(phi))
    y = center[1] + radius * np.outer(np.sin(theta), np.sin(phi))
    z = center[2] + radius * np.outer(np.ones(np.size(theta)), np.cos(phi))

    # Küreyi çiz
    ax.plot_surface(x, y, z, color=color, alpha=alpha)

def translate_sphere(ax, radius=1, center=(0, 0, 0), translation=(0, 0, 0), color='r', alpha=0.6):
    # Öteleme işlemi
    translated_center = (center[0] + translation[0], center[1] + translation[1], center[2] + translation[2])
    
    # Öteleme sonrası küreyi çiz
    draw_sphere(ax, radius, translated_center, color=color, alpha=alpha)

def main():
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Kullanıcıdan merkez koordinatlarını ve öteleme miktarını al
    center_x = float(input("Merkez X koordinatını girin: "))
    center_y = float(input("Merkez Y koordinatını girin: "))
    center_z = float(input("Merkez Z koordinatını girin: "))
    
    translation_x = float(input("X ekseninde öteleme miktarını girin: "))
    translation_y = float(input("Y ekseninde öteleme miktarını girin: "))
    translation_z = float(input("Z ekseninde öteleme miktarını girin: "))

    # Orijinal küreyi çiz
    draw_sphere(ax, radius=1, center=(center_x, center_y, center_z))

    # Ötelenmiş küreyi çiz
    translate_sphere(ax, radius=1, center=(center_x, center_y, center_z), translation=(translation_x, translation_y, translation_z))

    # Eksen etiketleri
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    plt.show()

if __name__ == "__main__":
    main()


```
