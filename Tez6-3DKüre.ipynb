```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Topun parametreleri
radius = 1

# Theta ve phi alar
theta = np.linspace(0, 2*np.pi, 100)
phi = np.linspace(0, np.pi, 100)

# Theta ve phi'yi bir zgara haline getir
theta, phi = np.meshgrid(theta, phi)

# Topun x, y, z koordinatlar
x = radius * np.sin(phi) * np.cos(theta)
y = radius * np.sin(phi) * np.sin(theta)
z = radius * np.cos(phi)

# 3D figr olutur
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Topu iz
ax.plot_surface(x, y, z, color='blue', alpha=0.7, edgecolor='k')

# Plot'u gster
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def draw_sphere(ax, radius=1, center=(0, 0, 0), color='b', alpha=0.6):
    # Kre izimi iin theta ve phi deerleri
    theta = np.linspace(0, 2 * np.pi, 100)
    phi = np.linspace(0, np.pi, 100)

    # Kre yzey noktalarnn hesaplanmas
    x = center[0] + radius * np.outer(np.cos(theta), np.sin(phi))
    y = center[1] + radius * np.outer(np.sin(theta), np.sin(phi))
    z = center[2] + radius * np.outer(np.ones(np.size(theta)), np.cos(phi))

    # Kreyi iz
    ax.plot_surface(x, y, z, color=color, alpha=alpha)

def translate_sphere(ax, radius=1, center=(0, 0, 0), translation=(0, 0, 0), color='r', alpha=0.6):
    # teleme ilemi
    translated_center = (center[0] + translation[0], center[1] + translation[1], center[2] + translation[2])
    
    # teleme sonras kreyi iz
    draw_sphere(ax, radius, translated_center, color=color, alpha=alpha)

def main():
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Kullancdan merkez koordinatlarn ve teleme miktarn al
    center_x = float(input("Merkez X koordinatn girin: "))
    center_y = float(input("Merkez Y koordinatn girin: "))
    center_z = float(input("Merkez Z koordinatn girin: "))
    
    translation_x = float(input("X ekseninde teleme miktarn girin: "))
    translation_y = float(input("Y ekseninde teleme miktarn girin: "))
    translation_z = float(input("Z ekseninde teleme miktarn girin: "))

    # Orijinal kreyi iz
    draw_sphere(ax, radius=1, center=(center_x, center_y, center_z))

    # telenmi kreyi iz
    translate_sphere(ax, radius=1, center=(center_x, center_y, center_z), translation=(translation_x, translation_y, translation_z))

    # Eksen etiketleri
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    plt.show()

if __name__ == "__main__":
    main()


```
