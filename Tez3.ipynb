```python
import numpy as np # sayısal işlemler ve çok boyutlu dizilerle çalışmak için kullanılır
import matplotlib.pyplot as plt

# Kullanıcıdan a, b ve c değerlerini alın
a = int(float(input("a değerini girin: ")))
b = int(float(input("b değerini girin: ")))
c = int(float(input("c değerini girin: ")))

# a^2 + b^2 ≠ 0 koşulunu kontrol edin
if a**2 + b**2 != 0:
    
    x = np.linspace(-10, 10, 400)  # -10 ile 10 arasındaki sayıları içeren ve 400 elemana sahip bir NumPy dizisidir.

    # Doğru denklemi
    y = ((-a * x) - c) / b # y yi yanlız bırakarak formülüze ettik

    # Doğruyu çizin
    plt.plot(x, y, label=f"{a}x + {b}y + {c} = 0") # label etiket

    # Eksenleri ayarlayın
    plt.axhline(0, color='black', linewidth=0.5) #eksenleri çizmek için kullanabileceğimiz bir kod
    plt.axvline(0, color='black', linewidth=0.5)
    
    # Öteleme miktarlarını kullanıcıdan alın
    x_oteleme = int(float(input("x ekseninde öteleme miktarını girin: ")))
    y_oteleme = int(float(input("y ekseninde öteleme miktarını girin: ")))

    # Ötelenmiş doğruyu çizin
    plt.plot(x + x_oteleme, y + y_oteleme, label=f"{x_oteleme}x kadar ve {y_oteleme}y kadar Ötelenmiş")

    # Grafik başlığını ve etiketleri ekleyin
    plt.title(f"{a}x + {b}y + ({c}) = 0")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend() # açıklama kutusu
    
    # Doğruyu göster
    plt.grid(True)
    plt.show()
else:
    print("a^2 + b^2 = 0 olmamalıdır. Lütfen geçerli değerler girin.")
```
