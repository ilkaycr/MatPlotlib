```python
import matplotlib.pyplot as plt   #matplotlib kütüphanesini ekleyip plt adıyla kullanıyoruz

def ciz_dogru(x1, y1, x2, y2):     #Doğru çekmek için fonksiyon yapabiliriz
    plt.plot([x1, x2], [y1, y2], 'b', label='Doğru Parçası')#Köşeli parantez başlangıç ve bitişi belirliyor 'b' blue mavi yapıyor çizgiyi  label='dogru' etikettir
    # A ve B noktalarını çizgi üzerine ekleyin
    plt.text(x1, y1, 'A', fontsize=12, ha='right', va='bottom')    #x1,y1 noktasına A yazdırır font büyüklüğü 12 noktanın sağına ve altına
    plt.text(x2, y2, 'B', fontsize=12, ha='right', va='bottom')

def otele_dogru(x1, y1, x2, y2, otele_x, otele_y):     #Doğruyu öteleme fonksiyonu
    x1 += otele_x
    x2 += otele_x
    y1 += otele_y
    y2 += otele_y
    plt.plot([x1, x2], [y1, y2], 'r', label='Ötelenmiş Doğru Parçası')
    # A ve B noktalarını çizgi üzerine ekleyin
    plt.text(x1, y1, 'A', fontsize=12, ha='right', va='bottom')
    plt.text(x2, y2, 'B', fontsize=12, ha='right', va='bottom')

# Kullanıcıdan çizginin başlangıç ve bitiş noktalarını alalım
x1 = float(input("X Başlangıç koordinatını girin: "))
x2 = float(input("X Bitiş koordinatını girin: "))
y1 = float(input("Y Başlangıç koordinatını girin: "))
y2 = float(input("Y Bitiş koordinatını girin: "))

# Doğruyu çiz fonksiyonunu kullanalım
ciz_dogru(x1, y1, x2, y2)

# Öteleme miktarını al
otele_x = float(input("X ekseni öteleme miktarını girin: "))
otele_y = float(input("Y ekseni öteleme miktarını girin: "))

# Doğruyu öteleme fonksiyonunu kullanalım
otele_dogru(x1, y1, x2, y2, otele_x, otele_y)


# Çizgileri görselleştir
plt.legend()  #Grafiklerin boş bir kısmına etiketleri yazdırır
plt.title('Doğru Parçası Öteleme')   #Başlık ekler
plt.xlabel('X Ekseni')       #xlabel yani x eksenine yazı ekler
plt.ylabel('Y Ekseni')       #ylabel yani x eksenine yazı ekler
plt.grid(False) #Kareleri açıp kapatmak için kullanabiliriz
plt.show() #Grafiği görmek için kullanırız

```
