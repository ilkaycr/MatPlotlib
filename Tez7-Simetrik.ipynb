```python
import matplotlib.pyplot as plt
import numpy as np

def simetri_bul(a, b, c, x, y):
    # Noktanın doğruya göre simetrisini bulma işlemi
    payda = a**2 + b**2
    
    # Koşul: payda sıfır olmamalı
    if payda != 0:
        simetri_x = x - ((2 * (a * x + b * y + c) * a) / payda)
        simetri_y = y - ((2 * (a * x + b * y + c) * b) / payda)
        return simetri_x, simetri_y
    else:
        return " a^2+b^2=0 --payda 0-- olamaz. ", None

# Kullanıcıdan doğru katsayılarını ve noktanın koordinatlarını al
a = float(input("Doğrunun a katsayısını girin: "))
b = float(input("Doğrunun b katsayısını girin: "))
c = float(input("Doğrunun c katsayısını girin: "))
x_koordinat = float(input("Noktanın X koordinatını girin: "))
y_koordinat = float(input("Noktanın Y koordinatını girin: "))

# Simetri işlemini uygula
sonuc = simetri_bul(a, b, c, x_koordinat, y_koordinat)

#tam sayı türünden sonuc yazdırmak için yapıyoruz
simetri_x_str = int(simetri_x)
simetri_y_str = int(simetri_y)
print(f"Simetri Noktası: ({simetri_x_str}, {simetri_y_str})")

# Doğruyu çiz
x = np.linspace(-10, 10, 100)
y_dogru = (-a * x - c) / b  #doğrunun eğrisini çizmek için kullanılır

# Simetri noktayı çiz

simetri_x, simetri_y = sonuc
plt.scatter([x_koordinat, simetri_x], [y_koordinat, simetri_y], color='red')


# Doğruyu ve noktayı çiz
plt.plot(x, y_dogru, label="Yansıma Ekseni")
plt.scatter(x_koordinat, y_koordinat, label="Nokta", color='blue')
plt.scatter(simetri_x, simetri_y, label="Simetri Nokta", color='red')

#noktaların koordinatlarını belirt
plt.text(x_koordinat, y_koordinat, f'({x_koordinat}, {y_koordinat})', fontsize=8, ha='right')
plt.text(simetri_x, simetri_y, f'({simetri_x}, {simetri_y})', fontsize=8, ha='right')

# Eksen etiketleri ve başlık
plt.xlabel('X Ekseni')
plt.ylabel('Y Ekseni')
plt.title('Noktanın Doğruya Göre Simetriği')

plt.legend() # Legand (açıklama) eklemek için
plt.grid(True) # Grid (ızgara) eklemek için
plt.show() # Grafik göster
```
