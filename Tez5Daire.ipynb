```python
import matplotlib.pyplot as plt
import numpy as np

# Kullanıcıdan girişleri al
yaricap = int(input("Dairenin yarıçapını girin: "))
renk1 = input("İlk dairenin rengini girin:red,yellow,green,black.. ")
x1 = int(input("İlk dairenin başlangıç noktasının x koordinatını girin: "))
y1 = int(input("İlk dairenin başlangıç noktasının y koordinatını girin: "))
renk2 = input("İkinci dairenin rengini girin:red,yellow,green,black.. ")
x_oteleme = int(input("Daireyi x ekseni boyunca öteleme miktarını girin: "))
y_oteleme = int(input("Daireyi y ekseni boyunca öteleme miktarını girin: "))

# Daire çizme fonksiyonu
def daire_ciz(yaricap, renk, x_oteleme=0, y_oteleme=0, etiket='', koordinat=''):
    theta = np.linspace(0, 2*np.pi, 100)
    
    x = yaricap * np.cos(theta) + x_oteleme
    y = yaricap * np.sin(theta) + y_oteleme
    
    plt.fill(x, y, color=renk)  #şeklin içini boyamak için kullanılır

    # Dairenin ortasındaki nokta
    plt.plot([x_oteleme], [y_oteleme], marker='o', markersize=5, color='black')  #Xorta=r.cos(0)+Xöteleme=r+Xöteleme , Yorta = r.sin(0)+Yöteleme = Yöteleme
    
    # Noktanın etiketi ve koordinatları
    plt.text(x_oteleme, y_oteleme, f'{etiket} ({koordinat})', ha='right', va='bottom', color='black')

# İki daireyi çiz
plt.figure()

# İlk daireyi çiz
daire_ciz(yaricap, renk1, x1, y1, 'A', f'{x1}, {y1}')

# İkinci daireyi belirtilen x ve y öteleme miktarlarına göre çiz
daire_ciz(yaricap, renk2, x1 + x_oteleme, y1 + y_oteleme, 'B', f'{x1 + x_oteleme}, {y1 + y_oteleme}')

plt.gca().set_aspect('equal', adjustable='box')  # Eksenlerin oranını koru
plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np

# Kullanıcıdan girişleri al
yaricap = int(input("Dairenin yarıçapını girin: "))
renk1 = input("İlk dairenin rengini girin: ör:red,yellow,green,black... ")
x1 = int(input("İlk dairenin başlangıç noktasının x koordinatını girin: "))
y1 = int(input("İlk dairenin başlangıç noktasının y koordinatını girin: "))
renk2 = input("İkinci dairenin rengini girin:red,yellow,green,black.. ")
x_oteleme = int(input("Daireyi x ekseni boyunca öteleme miktarını girin: "))
y_oteleme = int(input("Daireyi y ekseni boyunca öteleme miktarını girin: "))

# Daire çizme fonksiyonu
def bos_daire_ciz(yaricap, renk, x_oteleme=0, y_oteleme=0, etiket='', koordinat=''):
    theta = np.linspace(0, 2*np.pi, 100)
    
    x = yaricap * np.cos(theta) + x_oteleme
    y = yaricap * np.sin(theta) + y_oteleme
    
    plt.plot(x, y, color=renk, marker='', linestyle='-', linewidth=2)  # İçi boş daire çizimi

    # Dairenin ortasındaki nokta
    plt.plot([x_oteleme], [y_oteleme], marker='o', markersize=5, color='black')
    
    # Noktanın etiketi ve koordinatları
    plt.text(x_oteleme, y_oteleme, f'{etiket} ({koordinat})', ha='right', va='bottom', color='black')

# İki içi boş daireyi çiz
plt.figure()

# İlk içi boş daireyi çiz
bos_daire_ciz(yaricap, renk1, x1, y1, 'A', f'{x1}, {y1}')

# İkinci içi boş daireyi belirtilen x ve y öteleme miktarlarına göre çiz
bos_daire_ciz(yaricap, renk2, x1 + x_oteleme, y1 + y_oteleme, 'B', f'{x1 + x_oteleme}, {y1 + y_oteleme}')

plt.gca().set_aspect('equal', adjustable='box')  # Eksenlerin oranını koru
plt.show()
```
