```python
import numpy as np
import matplotlib.pyplot as plt

def tepe_noktasi_bul(a, b, c):
    tepe_x = -b / (2 * a)
    print(tepe_x)
    tepe_y = a * tepe_x**2 + b * tepe_x + c
    print(tepe_y)
    return tepe_x, tepe_y

def otolenmis_tepe_nok_bul(a, b, c, m, n):
    tepe_ox = -b / (2 * a) - m
    print(tepe_ox)
    tepe_oy = (a * tepe_x**2 + b * tepe_x + c) - n
    print(tepe_oy)
    return tepe_ox, tepe_oy

# Kullanıcıdan a, b ve c değerlerini alın
a = int(input("a değerini girin: "))
b = int(input("b değerini girin: "))
c = int(input("c değerini girin: "))

# Kullanıcıdan öteleme miktarlarını alın
x_oteleme = int(input("x ekseni öteleme miktarını girin: "))
y_oteleme = int(input("y ekseni öteleme miktarını girin: "))
m = x_oteleme
n = y_oteleme

# Kullanıcıdan aralık isteyin
baslangic = int(input("Başlangıç noktasını girin: "))
bitis = int(input("Bitiş noktasını girin: "))

# x değerlerini belirtilen aralıkta oluşturun
x = np.linspace(baslangic, bitis, 400)

# Parabolun y değerlerini hesaplayın
y = a * x**2 + b * x + c

# Tepe noktasını hesaplayın
tepe_x, tepe_y = tepe_noktasi_bul(a, b, c)

# Ötelenmiş parabolün y değerlerini hesaplayın
y_otelendi = a * (x + x_oteleme)**2 + b * (x + x_oteleme) + c - y_oteleme

# Tepe noktasını ötelenmiş parabol için de hesaplayın
tepe_ox, tepe_oy = otolenmis_tepe_nok_bul(a, b, c, m, n)

# Parabolü çizin
plt.plot(x, y, label="Parabol")
# Tepe noktasını işaretleyin
plt.scatter(tepe_x, tepe_y, color='blue', marker='o', label='Tepe Noktası (Orjinal)')
# Ötelenmiş parabolü çizin
plt.plot(x, y_otelendi, label=f"Ötelenmiş Parabol")
# Ötelenmiş parabolün tepe noktasını işaretleyin
plt.scatter(tepe_ox, tepe_oy, color='red', marker='o', label='Tepe Noktası (Ötelenmiş)')

# Eksenleri ayarlayın
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Grafik başlığını ve etiketleri ekleyin
plt.title("Parabol ve Ötelenmiş Parabol")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()

# y eksenini sınırlayın
plt.ylim(-100, 100)

# Parabolü ve ötelenmiş parabolü göster
plt.grid()
plt.show()
```
